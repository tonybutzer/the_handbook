```python
def nc_from_xarray(DS, filename):
    DS.time.attrs = {}  #this allowed the nc to be written
    #DS.SCL.attrs = {}
    ds1 = DS.drop(labels='spatial_ref')
    ds1.to_netcdf(filename)
```
