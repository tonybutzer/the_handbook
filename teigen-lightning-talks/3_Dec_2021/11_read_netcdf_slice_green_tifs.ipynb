```python
from datacube.utils.cog import write_cog
def bunch_of_tifs_from_xarray(DS, file_prefix):
    for i in range(0,len(DS)):
        # print(i)
        my_simple_date = str(DS[i].time.values).split(' ')[0]
        print(my_simple_date)
        tif_file_name = f'{file_prefix}_{my_simple_date}.tif'
        write_cog(DS[i], tif_file_name, overwrite=True)
```

```python
# help(write_cog) - overwrite=True added above
```

```python
%%time
bunch_of_tifs_from_xarray(rds_green,'siouxFalls_green')
```
