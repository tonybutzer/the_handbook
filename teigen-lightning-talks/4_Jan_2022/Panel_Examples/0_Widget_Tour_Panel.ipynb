```python
import panel as pn
pn.extension('tabulator')
pn.extension('terminal')
pn.extension()
```

```python
pn.Row('# A', '# B', '# C', background='lightgray')
```

```python
pn.Column('# 1', '# 2', '# 3', background='pink')
```

```python
pn.Column(
    pn.Row('# A', '# B', '# C', '## Markdown CSS', '#### you sunk my battleship',background='yellow'),
    '# 2', 
    '# 3', 
    background='pink')
```

```python
pn.Column(
    pn.Row('# .....','# A', '# B', '# C', '# D',background='lightgreen'),
    '# 1',
    '# 2', 
    '# 3',
    '# 4',
    background='cyan')
```

```python
widget = pn.widgets.TextInput(name='A widget', value='A string')
widget
```

```python
widget.value
```

```python
import random
rcolor = lambda: "#%06x" % random.randint(0, 0xFFFFFF)

box = pn.GridBox(*[pn.pane.HTML(background=rcolor(), width=50, height=50) for i in range(24)], ncols=6)
box
```

```python
pn.indicators.Dial(name='Failure Rate', value=10, bounds=(0, 100))
```

```python
progress = pn.indicators.Progress(name='Progress', value=20, width=200, bar_color='warning')
progress
```

```python
import time
for i in range(0,10):
    time.sleep(1)
    progress.value = i*9
    
```

```python
import pandas as pd
import datetime as dt
df = pd.DataFrame({
    'int': [1, 2, 3],
    'float': [3.14, 6.28, 9.42],
    'str': ['A', 'B', 'C'],
    'bool': [True, False, True],
    'date': [dt.date(2019, 1, 1), dt.date(2020, 1, 1), dt.date(2020, 1, 10)]
}, index=[1, 2, 3])

df_widget = pn.widgets.Tabulator(df)
df_widget
```
