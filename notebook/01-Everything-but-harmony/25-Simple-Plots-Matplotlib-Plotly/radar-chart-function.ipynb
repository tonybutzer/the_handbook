```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
from math import pi

%matplotlib inline  

def get_data():
    graph_settings = {'title': 'ODC Skills Assessment for Tony 2018',
                      'banners': ['ARCH','DEVOPS','PYTHON','MANAGEMENT','dataScience'],
                      'scores': [45,85,60,70,28]
                     }
    return graph_settings


def get_data_optimum():
    graph_settings = {'title': 'ODC Skills Optimum',
                      'banners': ['ARCH','DEVOPS','PYTHON','MANAGEMENT','dataScience'],
                      'scores': [90,90,90,90,90]
                     }
    return graph_settings


def graph_radar(graf):

    categories = graf['banners']
    N = len(categories)

    values = graf['scores']
    # We are going to plot the first line of the data frame.
    # But we need to repeat the first value to close the circular graph:
    values += values[:1]

    print(categories)

    print (values)
 
    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
 
    # Initialise the spider plot
    fig=plt.figure()


    ax = fig.add_subplot(111, polar=True)
 
    # Draw one axe per variable + add labels labels yet
    plt.xticks(angles[:-1], categories, color='white', size=8)
 
    # Draw ylabels
    ax.set_rlabel_position(0)
    plt.yticks([25,50,75], ["25","50","75"], color="grey", size=7)
    plt.ylim(0,100)

    title = graf['title']
    ax.set_title(title, color='yellow')


    # Plot data
    #ax.plot(angles, values, linewidth=1, linestyle='solid')
    ax.plot(angles, values, 'o-', linewidth=2)
 
    # Fill area
    ax.fill(angles, values, 'b', alpha=0.1)
    return True
```

```python
graf = get_data();
graph_radar(graf)
```

```python
graf = get_data_optimum();
graph_radar(graf)
```
