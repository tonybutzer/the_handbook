```python
# Import the library
import matplotlib.pyplot as plt
from matplotlib_venn import venn3
from matplotlib_venn import venn2

%matplotlib inline  
 
# Make the diagram
venn3(subsets = (10, 8, 22, 6,9,4,2))
plt.show()
```

```python
import pylab as plt
from matplotlib_venn import venn3, venn3_circles

v = venn3(subsets=(1,1,0,1,0,0,0))
v.get_label_by_id('100').set_text('First')
v.get_label_by_id('010').set_text('Second')
v.get_label_by_id('001').set_text('Third')
plt.title("Not a Venn diagram")
plt.show()
```

```python
set1 = set(['A', 'B', 'C', 'D'])
set2 = set(['B', 'C', 'D', 'E'])
set3 = set(['C', 'D',' E', 'F', 'G'])

venn3([set1, set2, set3], ('Set1', 'Set2', 'Set3'))
plt.show()
```

```python
erosl = ['dish','landsat','python','se','aws','constraint','satellite','git','c','oracle','cassandra','docker','lustre',
         'cogs']
gal = ['dish','landsat','python','scrum','aws','freedom','git','psql','docker','lustre','cogs']


eros = set(erosl)
ga = set(gal)


venn2([ga, eros], ('GA', 'EROS'))
plt.show()
```
