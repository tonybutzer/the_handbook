```python
# OLD approach
'''
import rasterio
tif_file = '/vsis3/ga-et-data-west/inputsv0/air_temperature/tavg_001.tif'
src = rasterio.open(tif_file)
src
'''

# new Nathan Roberts inspired method

import rasterio

tif_file = '/vsis3/ga-et-data/inputsv0/air_temperature/tavg_001.tif'


with rasterio.open(tif_file) as src:
    oviews = src.overviews(1) # list of overviews from biggest to smallest
    oview = oviews[-1] # let's look at the smallest thumbnail
    myThumb = src.read(1, out_shape=(1, int(src.height // oview), int(src.width // oview)))
```

```python
myThumb.shape
```

```python
from rasterio.plot import show_hist
show_hist(myThumb, bins=50, lw=0.0, stacked=False, alpha=0.3, histtype='stepfilled', title="Histogram")
```

```python
%matplotlib inline
from matplotlib import pyplot
pyplot.imshow(myThumb, cmap='gray')

pyplot.show()  
```

```python
with rasterio.open(tif_file) as src:
    oviews = src.overviews(1) # list of overviews from biggest to smallest
    oview = oviews[-2] # let's look at the 2nd smallest thumbnail
    myThumb = src.read(1, out_shape=(1, int(src.height // oview), int(src.width // oview)))

print(myThumb.shape)
pyplot.imshow(myThumb, cmap='gray')

pyplot.show()  
```
