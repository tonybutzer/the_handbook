```python
import boto3
import json
```

```python
def read_s3(s3_location):
    s3 = boto3.resource('s3')
    bucket, key = s3_location.split('/', 1)
    obj = s3.Object(bucket, key)
    body = str(obj.get()['Body'].read())
    print(type(body))
    return body

def get_work_records_s3(s3_location):
    body = read_s3(s3_location)
    work_records = body.split('\n')
    return work_records


def call_main(msg):
    print(msg)
    s3_location = 'rhassan-dev-test/dance_card.txt'
    work_records = get_work_records_s3(s3_location)
    print(work_records)

```

```python
call_main('hello')
```
