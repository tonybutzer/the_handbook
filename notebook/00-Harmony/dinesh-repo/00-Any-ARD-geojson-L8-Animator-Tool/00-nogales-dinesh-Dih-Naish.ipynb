```python
import datetime
from litestacLib.ls_api import Litestac
```

```python
my_aoi_file = './AOI/nogales.geojson'

date_range = (datetime.datetime(2017,5,1), datetime.datetime(2018,10,30))
# date_range = (datetime(2013,7,6), datetime(2018,7,7))

# es_index = 'rwanda'
# es_type = 'cube'

lc = Litestac()

measurements = ('red','green','blue',)

# return a pandas data frame product id creation date and red
panda_df = lc.search(my_aoi_file,
             time = date_range, measurements = measurements )
```

```python
pr_list=[]
for i,row in panda_df.iterrows():
    #print(row['path'], row['row'])
    pr= row['path'] + '-' +row['row']
    print(pr)
    pr_list.append(pr)
```

```python
def unique(list1): 
  
    # intilize a null list 
    unique_list = [] 
      
    # traverse for all elements 
    for x in list1: 
        # check if exists in unique_list or not 
        if x not in unique_list: 
            unique_list.append(x) 
    # print list 
    for x in unique_list: 
        print (x,)
    return(unique_list)
```
